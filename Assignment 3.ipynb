Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.google.com/maps")
```

```python
driver.maximize_window()
```

```python
searchbar=driver.find_element(By.ID,'searchboxinput')
searchbar.send_keys("Lucknow")
```

```python
searchbutton=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div[8]/div[3]/div[1]/div[1]/div/div[2]/div[1]/button')
searchbutton.click()
```

```python
try:
    url_string=driver.current_url
    print("URL Extracted :",url_string)
    lat_lng=re.findall(r'@(.*)data',url_string)
    if len(lat_lng):
        lat_lng_list = lat_lng[0].split(",")
        if len(lat_lng_list)>=2:
            lat = lat_lng_list[0]
            lng = lat_lng_list[1]
        print("Latitude = {}, Longitude = {}".format(lat, lng))

except Exception as e:
        print("Error: ", str(e))
```
