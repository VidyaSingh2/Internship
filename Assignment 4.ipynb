Q1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
driver.maximize_window()
```

```python
Video_Name=[]
Rank=[]
Artist=[]
Upload_Date=[]
Views=[]
```

```python
video=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
for i in video:
    Video_Name.append(i.text)
    
artist=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
for i in artist:
    Artist.append(i.text)
    
date_tag=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
for i in date_tag:
    Upload_Date.append(i.text)
    
views=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
for i in views:
    Views.append(i.text)
                          
```

```python
print(len(Video_Name),len(Views),len(Artist),len(Upload_Date))
```

```python
df=pd.DataFrame({"Video Name":Video_Name,"Artist":Artist,"Upload Date":Upload_Date,"Number of Views":Views})
df
```

Q4:Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used
Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://github.com/")
driver.maximize_window()
```

```python
open_source=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button')
open_source.click()
try:
    trending=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a')
    trending.click()
except ElementNotInteractableException as e:
    print("Element not visible")
```

```python
Rep_title=[]
Rep_desc=[]
Contributors=[]
Language=[]
url=[]
```

```python
url_tag=driver.find_elements(By.XPATH,'//h2[@class="h3 lh-condensed"]/a')
for i in url_tag:
    url.append(i.get_attribute("href"))
```

```python
len(url)
```

```python
for i in url:
    driver.get(i)
    time.sleep(3)
    
    try:
        title=driver.find_elements(By.XPATH,'//div[@id="repository-container-header"]/div[1]/div[1]/div/strong/a')
        for i in title:
            Rep_title.append(i.text)
    except NoSuchElementException as e:
        Rep_title.append("_")
        
    try:
        desc=driver.find_elements(By.XPATH,'//p[@class="f4 my-3"]')
        for i in desc:
            rep=i.text
            if rep is None:
                Rep_desc.append("-")
            else:
                Rep_desc.append(rep)
    except NoSuchElementException as e:
        Rep_desc.append("--")
    try:
        contributor=driver.find_elements(By.XPATH,'//div[@class="Layout-sidebar"]/div[1]/div[5]/div/h2/a/span')
        for i in contributor:
            Contributors.append(i.text)
    except NoSuchElementException as e:
        Contributors.append("-")
    
   
```

```python
language=driver.find_elements(By.XPATH,'//div[@class="f6 color-fg-muted mt-2"]/span[1]/span[2]')
for i in language:
    lang=i.text
    if lang is None:
        Language.append("-")
    else: 
        Language.append(lang)
```

```python
len(Language)
```

```python
Contributors.clear()
```

```python
Rep_desc
```

```python
print(len(Rep_title),len(Rep_desc),len(Contributors),len(Language))
```

```python
df=pd.DataFrame({"Repository Name":Rep_title,"Repository Description":Rep_desc,"Contributors Count":Contributors,"Language":Language})
df
```

Q8:Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/ You
have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute G) Year 
Note: - from the home page you have to go to the Show All Dataset page through code

```python
driver.get("https://archive.ics.uci.edu/")
```

```python
driver.maximize_window()
```

```python
Dataset_Name=[]
Data_Type=[]
Task=[]
Attribute_Type=[]
Instance_no=[]
Attribute_no=[]
Year=[]
```

```python
view_datasets=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/header/nav/ul/li[1]/a')
view_datasets.click()
```

```python
for page in range(0,664):
    expand_all=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]')
    expand_all.click()
    
    
    ds_name=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
    for i in ds_name:
        Dataset_Name.append(i.text)
    
    dtype=driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[2]/span')
    for i in dtype:
        Data_Type.append(i.text)
    
    
    task=driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[1]/span')
    for i in task:
        Task.append(i.text)
    
    att_type=driver.find_elements(By.XPATH,'//table[@class="col-span-full my-2 table sm:col-start-2"]/tbody/tr/td[2]')
    for i in att_type:
        Attribute_Type.append(i.text)
    
    instance_no=driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[3]/span')
    for i in instance_no:
        Instance_no.append(i.text)
    
    
    att_no=driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[4]/span')
    for i in att_no:
        Attribute_no.append(i.text)
    
    year=driver.find_elements(By.XPATH,'//table[@class="col-span-full my-2 table sm:col-start-2"]/tbody/tr/td[3]')
    for i in year:
        Year.append(i.text) 
    
    next_button=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')
    next_button.click()
    
    time.sleep(5)
```

```python
print(len(Dataset_Name),len(Data_Type),len(Task),len(Attribute_Type),len(Instance_no),len(Attribute_no),len(Year))
```

```python
df=pd.DataFrame({"Dateset Name":Dataset_Name,"Data Type":Data_Type,"Task":Task,"Attribute Type":Attribute_Type,"Instance Count":Instance_no,"Attribute Count":Attribute_no,"Year":Year})
df
```
